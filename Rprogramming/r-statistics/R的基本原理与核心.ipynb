#### 因子向量的建立

一个因子不仅包括分类变量本身，还包括变量不同的可能水平(**即使在数据中不出现**)。

```R
    factor(x, levels = sort(unique(x)), na.last = TRUE),
    labels = levels, exclude = NA, ordered = is.ordered(x))
```

levels用来指定因子的水平，labels用来指定水平的名字；exclude表示从向量x中剔除的水平值；ordered是一个逻辑选项，用来指定因子的水平是否有次序。

````r
# 将字符型向量转换为因子
a = c("green", "blue", "green", "yellow")
a = factor(a)
a
str(a) # 本来在R或Rstudio可以直接看到a输出结果的因子表示，但jupyter notebook不支持显示，所以我用str()函数将其显示出来，这样便于大家理解
````

````r
# 将数值型向量转换为因子
b = c(1, 2, 3, 1)
b = factor(b)
b
str(b)
````

````r
# 将字符型因子转换为数值型因子
a = c("green", "blue", "green", "yellow")
a = factor(a)
levels(a) = c(1, 2, 3, 4)
a
str(a)
````

````r
ff = factor(c("A", "B", "C"), labels = c(1, 2, 3))
ff
str(ff)
````

````r
# 将数值型因子转换为字符型因子
b = c(1, 2, 3, 1)
b = factor(b)
levels(b) = c("low", "middle", "high")
b
str(b)
````

````r
ff = factor(c(2,4), levels=2:5)
ff
str(ff)
levels(ff)
````

````r
# 函数gl()能产生规则的因子序列，用法为gl(k,n)，其中k是水平数，n是每个水平重复次数。length选项指定产生数据的个数，label指定水平因子的名字
gl(3, 5)
str(gl(3,5))
````

默认函数的使用是按列计算的，也可以通过MARGIN选项改变。

apply()函数可以对一个对象施加某种运算函数，调用格式：

```R
apply(X, MARGIN, FUN)
```
X为参与运算矩阵，FUN为一个函数，MARGIN=1表示按行计算；2表示按列计算。

另有sweep()函数

```R
sweep(X, MARGIN, STATS, FUN)
```

表示从矩阵X中按MARGIN计算STATS，并从X中除去

下面举例：

````r
a = matrix(rnorm(n=12), nr=3)
````

````r
a
````

````r
# 求平均数、中位数等
apply(a, MARGIN=1, FUN=mean) #求各行的均值
````

````r
apply(a, MARGIN=2, FUN=mean) #求各列的均值
````

````r
# 标准化
scale(a, center=T, scale=T)
````

````r
# 减去中位数
row.med = apply(a, MARGIN=1, FUN=median)
sweep(a, MARGIN=1, STATS=row.med, FUN="-")
````

### 时间序列的建立

函数ts()通过一向量或者矩阵创建一个一元或多元的时间序列，称为ts型对象。

调用格式：

```R
ts(data = NA, start = 1, end = numeric(0), frequency = 1, 
    deltat = 1, ts.eps = getOption("ts.eps"), class, names)
```

选项：

data —— 一个向量或者矩阵

start —— 第一个观察值的时间

end —— 最后一个观察值的时间

frequency —— 单位时间内观察值的频数

deltat —— 两个观察值的时间间隔，要么给定这个选项，要么给定frequency

ts.eps —— 序列之间的误差限

...

````r
# 下面举例说明
ts(1:10, start = 1959)
````

````r
ts(1:47, frequency=12, start=c(1959, 2))
````

````r
ts(1:10, frequency=4, start=c(1959, 2))
````

````r
ts(matrix(rpois(36, 5), 12, 3), start=c(1961,1), frequency=12)
````

## R基本绘图 

R提供了丰富的绘图功能，我们可以通过R提供的示例进行了解

````r
# 二维图形示例
demo(graphics)
````

````r
# 三维图形示例
demo(persp)
````

**绘图函数的工作方式与本文前面描述的工作方式大为不同，不能把绘图函数的结果赋给一个对象，其结果将直接输出到一个“绘图设备”上。绘图设备是一个绘图的窗口或时一个文件。**

R有两种绘图函数：

- 高级绘图函数： 创建一个新的图形
- 低级绘图函数： 在现存的图形上添加元素

另外绘图参数提供了丰富的绘图选项，可以使用缺省值或者用函数`par()`修改。更高级的图形使用`grid`和`lattice`绘图包实现，而近几年`ggplot`包是绘图一大热点。


绘图函数众多，我不一一列出了，靠死记是掌握不了技能的，一些基本图形不会可以百度谷歌找资料，像https://www.harding.edu/fmccown/r/ 就展示了一些简单图形的画法。我们接下来也以实际操作，使用的命令为常用命令。

### 一个实例

以内嵌数据Puromycin为例来说明R中基本的绘图方法

````r
dim(Puromycin)
````

````r
head(Puromycin)
````

````r
# 简单散点图
# 对于状态treated，画出rate关于cone的散点图
PuroA = subset(Puromycin, state == "treated")
plot(rate ~ conc, data = PuroA)
````

有三种方法指明函数`plot()`使用的数据集

- `plot()`函数中使用data选项

- 在`with()`中使用`plot()`
    ```R
    with(PuroA, plot(conc, rate))
    ```

- 使用`$`直接指向数据与变量

````r
# plot(PuroA$rate, PuroA$conc) 
````

#### 美化图形

````r
# R提供25种不同的符号和8种不同颜色，浏览它们的命令是
u = 1:25
plot(u ~ 1, pch = u, col = u, cex = 3)
````

````r
# 选择合适的符号及其大小与颜色
plot(rate ~ conc, data = PuroA, pch = 2, col = 4, cex = 2.5)
````

````r
# 坐标轴与标题设定
plot(rate ~ conc, data = PuroA, pch = 2, col = 4,
    cex = 2.5, xlim = c(0, 1.2), ylim = c(40, 210),
    ylab = "Concentration",
    xlab = "Rate", cex.lab = 2)
title(main = "Puromycin", cex.main = 3)
````
